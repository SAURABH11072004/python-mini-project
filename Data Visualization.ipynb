```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the CSV file

data = pd.read_csv("E:\mini project\HDFC.csv")

# Convert timestamp to datetime for easier handling of time-based data
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Check data for any preprocessing needs
print(data.info())
```

```python
#HISTOGRAM
plt.figure(figsize=(10, 6))
sns.histplot(data['close'], bins=30, kde=True, color='blue')
plt.title('Histogram of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar Plot
data['hour'] = data['timestamp'].dt.hour
hourly_avg_close = data.groupby('hour')['close'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=hourly_avg_close.index, y=hourly_avg_close.values, palette="viridis")
plt.title('Average Closing Price by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Closing Price')
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['close'], color="orange")
plt.title('Box Plot of Closing Prices')
plt.xlabel('Closing Price')
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x=data['close'], color="purple")
plt.title('Violin Plot of Closing Prices')
plt.xlabel('Closing Price')
plt.show()
```

```python
#Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x=data['hour'], palette="coolwarm")
plt.title('Count of Data Points by Hour')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='close', y='volume', data=data, color='teal')
plt.title('Scatter Plot of Closing Price vs Volume')
plt.xlabel('Closing Price')
plt.ylabel('Volume')
plt.show()
```

```python
#Line Plot
plt.figure(figsize=(14, 6))
sns.lineplot(x='timestamp', y='close', data=data, color='darkblue')
plt.title('Line Plot of Closing Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Joint Plot
sns.jointplot(x='close', y='volume', data=data, kind="scatter", color="g")
plt.suptitle('Joint Plot of Closing Price vs Volume', y=1.02)
plt.show()
```

```python
#Pair plot
sns.pairplot(data[['open', 'high', 'low', 'close', 'volume']])
plt.suptitle('Pair Plot of Financial Data', y=1.02)
plt.show()
```

```python
#HeatMap
plt.figure(figsize=(10, 6))
sns.heatmap(data[['open', 'high', 'low', 'close', 'volume']].corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#RugPlot
plt.figure(figsize=(10, 2))
sns.rugplot(data['close'], height=0.1, color='green')
plt.title('Rug Plot of Closing Prices')
plt.xlabel('Closing Price')
plt.show()
```
